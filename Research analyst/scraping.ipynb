```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_schools_in_Maharashtra'
data = requests.get(url)
data
```

```python
bs = BeautifulSoup(data.content)
```

```python
headings_tags = bs.find_all('span',class_="mw-headline")
headings_tags
```
